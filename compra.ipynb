```python
import numpy as np
import random as rd
import scipy.stats as sst
import matplotlib.pyplot as plt
```

```python
def demanda():
  compra = [100,200,300,400,500,600]
  Mes=[8,9,4,9,4,2]
  n=len(Mes)
  a = rd.random()

  for x in range(n):
    if a>sum(Mes[0:x])/sum(Mes):
      if a<=sum(Mes[0:x+1])/sum(Mes):
        d = compra[x]
        break
  E=[]
  for x1 in range(n):
    if compra[x1]<=d:
      U = compra[x1]*(1000-750)
    else:
      U = d*1000 - compra[x1]*750 + (compra[x1]-d)*250

    E.append(U) 
  return(compra[E.index(max(E))])
```

```python
C=demanda()
```

```python
def compra_promedio(n):
  l1 =[]
  l2 =[] 
  for x in range(n):
    l1.append(demanda())
    l2.append(np.mean(l1))
  return(l2)
```

```python
c_p = compra_promedio(200)

print(np.mean(c_p))
plt.plot(c_p)
plt.show()
```
